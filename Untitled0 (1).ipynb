```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns

# ============ DATASET ============
np.random.seed(42)
n = 5000

canais  = ['Telefone', 'Chat', 'E-mail', 'WhatsApp']
turnos  = ['Manhã', 'Tarde', 'Noite']
motivos = ['Dúvida sobre plano', 'Cancelamento', 'Reembolso',
           'Agendamento de consulta', 'Resultado de exame',
           'Cobrança indevida', 'Credenciamento', 'Urgência médica']
status  = ['Resolvido', 'Pendente', 'Transferido']
agentes = [f'Agente_{i:02d}' for i in range(1, 21)]

tma_base = {
    'Telefone': (360, 120), 'Chat': (180, 60),
    'E-mail': (480, 150),   'WhatsApp': (240, 90),
}

canal  = np.random.choice(canais, n, p=[0.45, 0.25, 0.15, 0.15])
turno  = np.random.choice(turnos, n, p=[0.40, 0.38, 0.22])
motivo = np.random.choice(motivos, n, p=[0.20, 0.08, 0.12, 0.18, 0.15, 0.10, 0.09, 0.08])
agente = np.random.choice(agentes, n)

tma = np.array([max(60, int(np.random.normal(tma_base[c][0], tma_base[c][1]))) for c in canal])
tma[motivo == 'Urgência médica']   = (tma[motivo == 'Urgência médica'] * 1.5).astype(int)
tma[motivo == 'Cancelamento']      = (tma[motivo == 'Cancelamento'] * 1.3).astype(int)
tma[motivo == 'Cobrança indevida'] = (tma[motivo == 'Cobrança indevida'] * 1.2).astype(int)
tma[turno == 'Noite']              = (tma[turno == 'Noite'] * 1.15).astype(int)

datas = pd.to_datetime(np.random.choice(pd.date_range('2023-01-01', '2023-12-31'), n))

df = pd.DataFrame({
    'data': datas, 'canal': canal, 'turno': turno,
    'agente': agente, 'motivo': motivo,
    'tma_segundos': tma,
    'status': np.random.choice(status, n, p=[0.72, 0.18, 0.10])
})
df['tma_minutos'] = (df['tma_segundos'] / 60).round(1)
df['dia_semana']  = df['data'].dt.day_name()
df['mes']         = df['data'].dt.month
df['hora']        = np.random.choice(range(7, 22), n)
df['faixa_tma']   = pd.cut(df['tma_minutos'], bins=[0,3,6,10,999],
                            labels=['Rápido','Normal','Lento','Crítico'])

print(f"Dataset criado: {df.shape}")

# ============ GRÁFICOS ============
COR_PRINCIPAL  = '#1A6B8A'
COR_DESTAQUE   = '#E84B3A'
COR_SECUNDARIA = '#2EAF8E'
COR_AMARELO    = '#F0A500'
COR_FUNDO      = '#F8F9FA'
COR_TITULO     = '#1A1A2E'
COR_TEXTO      = '#4A4A6A'

plt.rcParams.update({
    'font.family': 'DejaVu Sans',
    'axes.spines.top': False, 'axes.spines.right': False,
    'axes.spines.left': False, 'axes.spines.bottom': False,
    'axes.grid': True, 'grid.color': '#E8E8E8', 'grid.linewidth': 0.8,
    'xtick.color': COR_TEXTO, 'ytick.color': COR_TEXTO,
    'xtick.labelsize': 10, 'ytick.labelsize': 10,
})

fig = plt.figure(figsize=(18, 14), facecolor=COR_FUNDO)
fig.text(0.5, 0.98, 'Análise de Central de Atendimento — TMA e Volume 2023',
         ha='center', fontsize=18, fontweight='bold', color=COR_TITULO)
fig.text(0.5, 0.955, 'Tempo Médio de Atendimento, volume por canal, turno, motivo e agente',
         ha='center', fontsize=12, color=COR_TEXTO)

axes = fig.subplots(3, 2)
for ax in axes.flat:
    ax.set_facecolor(COR_FUNDO)

ax1 = axes[0, 0]
tma_canal = df.groupby('canal')['tma_minutos'].mean().sort_values(ascending=False)
cores1 = [COR_DESTAQUE if v == tma_canal.max() else COR_PRINCIPAL for v in tma_canal.values]
bars1 = ax1.barh(tma_canal.index, tma_canal.values, color=cores1, height=0.5)
ax1.set_title('TMA médio por canal (minutos)', fontsize=12, fontweight='bold', color=COR_TITULO, pad=12)
ax1.axvline(tma_canal.mean(), color='#AAAAAA', linestyle='--', linewidth=1)
ax1.text(tma_canal.mean()+0.1, -0.6, f'Média: {tma_canal.mean():.1f}min', fontsize=8, color='#AAAAAA')
for bar, val in zip(bars1, tma_canal.values):
    ax1.text(val+0.1, bar.get_y()+bar.get_height()/2, f'{val:.1f}min', va='center', fontsize=9, fontweight='bold', color=COR_TITULO)

ax2 = axes[0, 1]
vol_canal = df['canal'].value_counts()
cores2 = [COR_DESTAQUE if v == vol_canal.max() else COR_SECUNDARIA for v in vol_canal.values]
bars2 = ax2.bar(vol_canal.index, vol_canal.values, color=cores2, width=0.5)
ax2.set_title('Volume de atendimentos por canal', fontsize=12, fontweight='bold', color=COR_TITULO, pad=12)
ax2.set_ylim(0, vol_canal.max()*1.2)
for bar, val in zip(bars2, vol_canal.values):
    pct = val/len(df)*100
    ax2.text(bar.get_x()+bar.get_width()/2, val+20, f'{val}\n({pct:.0f}%)', ha='center', fontsize=9, fontweight='bold', color=COR_TITULO)

ax3 = axes[1, 0]
tma_motivo = df.groupby('motivo')['tma_minutos'].mean().sort_values(ascending=False)
cores3 = [COR_DESTAQUE if v >= tma_motivo.quantile(0.75) else COR_PRINCIPAL for v in tma_motivo.values]
bars3 = ax3.barh(tma_motivo.index, tma_motivo.values, color=cores3, height=0.5)
ax3.set_title('TMA médio por motivo de contato', fontsize=12, fontweight='bold', color=COR_TITULO, pad=12)
ax3.set_xlabel('Minutos', fontsize=10, color=COR_TEXTO)
for bar, val in zip(bars3, tma_motivo.values):
    ax3.text(val+0.1, bar.get_y()+bar.get_height()/2, f'{val:.1f}min', va='center', fontsize=9, fontweight='bold', color=COR_TITULO)

ax4 = axes[1, 1]
ordem_dias  = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
nomes_dias  = ['Seg','Ter','Qua','Qui','Sex','Sáb','Dom']
vol_dia = df['dia_semana'].value_counts().reindex(ordem_dias).fillna(0)
cores4 = [COR_DESTAQUE if v == vol_dia.max() else COR_PRINCIPAL for v in vol_dia.values]
bars4 = ax4.bar(nomes_dias, vol_dia.values, color=cores4, width=0.5)
ax4.set_title('Volume de atendimentos por dia da semana', fontsize=12, fontweight='bold', color=COR_TITULO, pad=12)
ax4.set_ylim(0, vol_dia.max()*1.2)
for bar, val in zip(bars4, vol_dia.values):
    ax4.text(bar.get_x()+bar.get_width()/2, val+10, str(int(val)), ha='center', fontsize=9, fontweight='bold', color=COR_TITULO)

ax5 = axes[2, 0]
tma_turno = df.groupby('turno')['tma_minutos'].mean().sort_values(ascending=False)
cores5 = [COR_DESTAQUE if v == tma_turno.max() else COR_AMARELO for v in tma_turno.values]
bars5 = ax5.bar(tma_turno.index, tma_turno.values, color=cores5, width=0.4)
ax5.set_title('TMA médio por turno', fontsize=12, fontweight='bold', color=COR_TITULO, pad=12)
ax5.set_ylabel('Minutos', fontsize=10, color=COR_TEXTO)
ax5.set_ylim(0, tma_turno.max()*1.25)
for bar, val in zip(bars5, tma_turno.values):
    ax5.text(bar.get_x()+bar.get_width()/2, val+0.1, f'{val:.1f}min', ha='center', fontsize=11, fontweight='bold', color=COR_TITULO)

ax6 = axes[2, 1]
top_agentes   = df.groupby('agente')['tma_minutos'].mean().sort_values(ascending=False).head(10).index
df_top        = df[df['agente'].isin(top_agentes)]
ordem_agentes = df_top.groupby('agente')['tma_minutos'].mean().sort_values(ascending=False).index
bp = ax6.boxplot([df_top[df_top['agente']==ag]['tma_minutos'].values for ag in ordem_agentes],
                  labels=ordem_agentes, vert=False, patch_artist=True,
                  boxprops=dict(facecolor=COR_PRINCIPAL, alpha=0.6),
                  medianprops=dict(color=COR_DESTAQUE, linewidth=2),
                  whiskerprops=dict(color=COR_TEXTO),
                  capprops=dict(color=COR_TEXTO),
                  flierprops=dict(marker='o', color=COR_DESTAQUE, markersize=3))
ax6.set_title('Distribuição de TMA — Top 10 agentes\n(linha vermelha = mediana)',
              fontsize=12, fontweight='bold', color=COR_TITULO, pad=12)
ax6.set_xlabel('Minutos', fontsize=10, color=COR_TEXTO)

fig.text(0.5, 0.01, 'Fonte: dados simulados com base em padrões reais de call center  •  Análise: Python / pandas / matplotlib',
         ha='center', fontsize=9, color='#AAAAAA')

plt.tight_layout(rect=[0, 0.02, 1, 0.94])
plt.savefig('analise_central_atendimento.png', dpi=180, bbox_inches='tight', facecolor=COR_FUNDO)
plt.show()
print("✅ Salvo com sucesso!")
```
